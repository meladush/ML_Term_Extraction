```python
from sentence_transformers import SentenceTransformer, util
st_model = SentenceTransformer("distiluse-base-multilingual-cased-v2")
```

Отбираем случайные 50 предложений, чтобы отложить их для валидационной выборки. Векторы эмбеддингов будем сохранять отдельно для обучающей и валидационной выборок.

```python
index_list = [685, 649, 301, 63, 626, 417, 551, 719, 381, 606, 271, 753, 397, 749, 634, 284, 222, 94, 126, 482, 355, 303, 363, 255, 264, 454, 25, 341, 173, 224, 292, 532, 475, 236, 5, 494, 345, 594, 721, 723, 752, 757, 424, 371, 452, 186, 18, 382, 653, 645]
```

```python
df_to_split = pd.read_json("/content/drive/MyDrive/ML_Data/train_t1_v1.jsonl", lines=True)
```

```python
sents = [df_to_split['text'][i] for i in index_list]
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML_Data/prerpocessed_for_embeddings.csv', sep=';', encoding='utf-8')
```

```python
df = df.drop('Unnamed: 0', axis=1).rename(columns={"sent": "text"})
```

```python
df
```

```python
def check(x):

  text = x['text']
  return text in sents
```

```python
df['check'] = df.apply(check, axis=1)
```

```python
df
```

```python
df_val = df.loc[df['check'] == True]
df_val
```

```python
df_val[['ngram', 'text', 'pos_tags', 'label']].to_csv('/content/drive/MyDrive/ML_Data/for_validation.csv', sep=';', encoding='utf-8')
```

```python
df_for_class = df.loc[df['check'] == False]
df_for_class
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
# для каждой строки датафрейма создаем вектор, конкатенирующий эмбединг n-граммы, эмбединг предложения, в котором
# она встречатся, а также уникальный код набора частеречных тегов. Все векторы объединяем в numpy-массив.

def get_vectors(dataframe):

  all_vectors = []

  ngrams = dataframe['ngram'].to_list()
  texts = dataframe['text'].to_list()
  pos_tags = np.asarray(dataframe['pos_tags'])
  pos_tags = le.fit_transform(pos_tags)

  for i in range(len(ngrams)):
    ngram_emb = st_model.encode(ngrams[i])
    text_emb = st_model.encode(texts[i])
    pos_tag = np.asarray([pos_tags[i]])

    vector = np.concatenate([ngram_emb, text_emb, pos_tag])
    all_vectors.append(vector)

  return np.asarray(all_vectors)
```

```python
def get_labels(dataframe):

  return np.asarray(dataframe['label'])
```

```python
# сохраняем валидационный датасет из 50 текстов
```
